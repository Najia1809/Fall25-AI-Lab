#  Data Pre-Processing
### Manage Null-values

```python
df.drop("salary_in_usd",axis=1,inplace=True)
df
```

```python
df['salary_currency'] =df['salary_currency'].fillna(df['salary_currency'].mode()[0])
```

```python
df
```

```python
df.columns
```

```python
def changefloattypes(cols):
    for i in cols:
        df[i] = df[i].astype('int64')

columns_to_int = ['work_year', 'salary', 'remote_ratio']
changefloattypes(columns_to_int)
```

```python
def changeonstringintoInt(cols):
    for i in cols:
        data_LE = LabelEncoder()
        data_LE.fit(np.ravel(df[i]))
        df[i] = data_LE.transform(df[i])

columns = ['experience_level', 'employment_type', 'job_title',
            'salary_currency', 'employee_residence',
            'company_location', 'company_size']

changeonstringintoInt(columns)
```
