## **Q: Write code for minmax alogrithm**

```python
values = [4,6,8,3,7,9,2,5]
def minmax(currlevel,index,isMax):
    if currlevel ==3:
        return values[index]
    if isMax:
        left = minmax(currlevel+1,index*2,False)
        right  = minmax(currlevel+1,index*2,False)
        return max(left,right)
    else:
        left = minmax(currlevel+1,index*2,True)
        right  = minmax(currlevel+1,index*2,True)
        return min(left,right)
result = minmax(0,0,True)
print(result)
```
