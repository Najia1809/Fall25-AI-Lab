# MiniProject 2:

```python
movies = [
("Eternal Sunshine of the Spotless Mind", 20000000), ("Memento", 9000000),
("Requiem for a Dream", 4500000),
("Pirates of the Caribbean: On Stranger Tides", 379000000), ("Avengers: Age of Ultron", 365000000),
("Avengers: Endgame", 356000000),
("Incredibles 2", 200000000)
]

new_movies = int(input("How many movies you want to enter: "))
for i in range(new_movies):
    name = input("enter movie name")
    budget = int(input("enter budget for movie"))
    movies.append((name,budget))

print("\nBudget of all movies: ")
budgets = [budget for name, budget in movies]
print(budgets)

print("\nAverage of budget: ")
average = sum(budgets)/len(budgets)
print(average)

print("\nMovies name that has budget greater then average: ")

count = 0
for name,budget in movies:
    if budget>average:
        differnce = budget-average
        print(f"{name},\nhigher then average by {differnce}")
        count +=1
print("\nTotal movies above average: ",count)
```
